```python
questions=["1. How many people have heart disease and how many don't have heart disease?",
           "2. People of which sex have most heart disease?",
           "3. People of which sex has which type of chest pain most?",
           "4. People with which chest pain are most pron to have heart disease?",
          "5. Peope of which age are most prone to heart disease?",
          "6. relationsship between  fbs and heart diseases",
          "7. average chol with respect to their age",
          ]
questions
```

```python
#1. How many people have heart disease and how many don't have heart disease?

#0 --> person not having heart disease
#1--> person having heart disease
df.target.value_counts()
```

```python
#plotting bar graph
df.target.value_counts().plot(kind='bar',color=["orchid","salmon"])
plt.title("Heart disease values")
plt.xlabel("1 = heart disease,0 = No heart disease")
plt.ylabel("Amount");
```

```python
#plot pie chart
df.target.value_counts().plot(kind="pie",figsize=(6,6))
plt.legend(["Disease","No disease"]);
```

```python
# Male female ratio
# 1-->Male  0-->female
# 0--> no disease  1-->disease
df.sex.value_counts()
```

```python
#Plotting pie chart
df.sex.value_counts().plot(kind='pie',figsize=(6,6))
plt.title("Male female ratio");
plt.legend(['Male','Female']);
```

```python
# 2. People of which sex have most heart disease?
#0-->female 1-->male
pd.crosstab(df.target,df.sex)
```

```python
sns.countplot(x='target',data=df,hue='sex')
plt.title("Heart disease frequency for sex")
plt.xlabel("0--> No heart disease,  1-->heart disease");
```

```python
# 3.People of which sex has which type of chest pain most

#counting values for different chest pain
df.cp.value_counts()
```

```python
df.cp.value_counts().plot(kind='bar',color=['salmon','lightskyblue','springgreen','khaki'])
plt.title("Chest pain type v/s count");
```

```python
pd.crosstab(df.sex,df.cp)
```

```python
pd.crosstab(df.sex,df.cp).plot(kind = 'bar',color=['coral','lightskyblue','plum','khaki'])
plt.title("Type of chest pain for sex")
plt.xlabel("0=Female,    1=Male");
```

```python
# 4. people with which chest pain are most pron to heart disease

pd.crosstab(df.cp,df.target)
```

```python
sns.countplot(x='cp',data=df,hue='target');
```

```python
# distribution plot with normal distribution curve
sns.displot(x='age',data=df,bins=30,kde=True);
```

```python
# distribution plot for 'maximum heart rate'

sns.displot(x='thalach',data=df,bins=30,kde=True,color='chocolate');
```

```python
# 5.people of which age are most prone to heart disease
df.age.value_counts()
```

```python
#plotting bar graph
df.age.value_counts().plot(kind='bar', figsize=(20,5),grid=False,color=["violet","indigo","blue","green","yellow","orange","red"])
plt.title("Age of peoples prone to heart disease")
plt.xlabel("Age")
plt.ylabel("Amount");
```

```python
pd.crosstab(df.age,df.target).plot(kind = 'bar',figsize=(20,5),grid=True)
plt.title("Different age of people prone to heart disease with different sex")
plt.xlabel("0=Female,    1=Male");
```

```python
# 6.relationsship between  fbs and heart diseases
#0--> dont'have   1-->have
pd.crosstab(df.fbs,df.target)
```

```python
sns.countplot(x='fbs',data=df,hue='target');
plt.xlabel("0=no fbs   ,    1=fbs");
```

```python
# distribution plot with normal distribution curve
sns.displot(x='fbs',data=df,bins=30,kde=True);
```

```python
# 7.average chol or people with respect to their age
df.groupby('age')['chol'].mean()
```

```python
age_chol_avg = df.groupby('age')['chol'].mean().reset_index()

# Plot the data
plt.figure(figsize=(20, 5))
plt.bar(age_chol_avg['age'], age_chol_avg['chol'],color=['skyblue','springgreen','chocolate'])
plt.xlabel('Age')
plt.ylabel('Average Cholesterol (chol)')
plt.title('Average Cholesterol with Respect to Age')
plt.grid(False)
plt.show()
```
